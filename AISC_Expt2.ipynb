```
import math
import numpy as np
from matplotlib import pyplot as plt
```

```
i1 = 0.8
i2 = 0.6
i3 = 0.4
w1 = 0.1
w2 = 0.3
w3 = -0.2
bias = 0.35
X = w1*i1 + w2*i2 + w3*i3 + bias
```

Hard Limit Transfer Function

```
def hard_limit(y):
  if y>=0:
    return 1
  else:
    return 0
```

```
Y=hard_limit(X)
print(Y)
```

Binary Step

```
def bin_step(x):
    threshold = 0 
    if(x>=threshold):
      return(1)
    else:
      return(0)
```

```
Y=bin_step(X)
print(Y)
```

Bipolar Step

```
def bip_step(x):
    threshold = 0 
    if(x>=threshold):
      return(1)
    else:
      return(-1)
```

```
Y=bip_step(X)
print(Y)
```

Sigmoid Function

```
def bin_sigmoid(x):
    return 1/(1+np.exp(-x))
```

```
Y=bin_sigmoid(X)
print(Y)
```

Saturating Linear Function


```
def saturating_linear(x):
  if x<=0:
    return 0
  elif x>0 and x<1:
    return x
  else:
    return 1
```

```
Y=saturating_linear(X)
print(Y)
```

Tansig Function


```
def tansig(x):
    return 2/(1+np.exp(-2*x))-1
```

```
Y=tansig(X)
print(Y)
```

Log Sigmoid Function

```
def logsigmoid(y):
  return 1/(1+math.exp(-y))
```

```
Y=logsigmoid(X)
print(Y)
```

```
ip = [[0,0],[0,1],[1,0],[1,1],[0.2,0.9],[0.9,0.2],[0.9,0.9],[0.3,0.3]]
op = [0,1,0,1,1,0,1,0]
```

```
theta = 0.5
lr = 0.3
w1 = 0 
w2 = 0
b = 0
```

```
epochs = 400
```

```
def calculate_y(y,theta):
  if y > theta:
    return 1
  else:
    return 0
 
```

```
for i in range(epochs):
  print("Epoch ",i)
  for j in range(len(ip)):
    yin = b + w1*ip[j][0] + w2*ip[j][1]
    y = calculate_y(yin,theta)
    print("calculated", y, op[j])
    if y != op[j]:
        w1 = w1 + lr*op[j]*ip[j][0]
        w2 = w2 + lr*op[j]*ip[j][1]
        b = b + lr*op[j]
        print("w1", w1, "w2", w2, "b", b)



    
```

```
for i in ip:
  print(i, end = " ")
  y_in = b + w1*i[0] + w2*i[1]
  if y_in > theta:
    print(1)
  else: print(0)
```

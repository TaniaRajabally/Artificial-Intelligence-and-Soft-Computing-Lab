```
#AND OR NOT 
def AND(x1,x2):
  w1 = 1
  w2 = 1
  b = -1
  if w1*x1 + w2*x2 + b > 0:
    return 1
  else:
    return 0

def OR(x1,x2):
  w1 = 2
  w2 = 2
  b = -1
  if w1*x1 + w2*x2 + b > 0:
    return 1
  else:
    return 0

def NOT(x):
  w = -1
  b = 1
  if w*x + b > 0:
    return 1
  else:
    return 0
```

Logical Expression for SUM:
= A’ B’ C-IN + A’ B C-IN’ + A B’ C-IN’ + A B C-IN



```
def SUM_FN(A,B,C):
  y1 = AND(AND(NOT(A),NOT(B)),C)
  y2 = AND(AND(NOT(A),B),NOT(C))
  y3 = AND(AND(A,NOT(B)),NOT(C))
  y4 = AND(AND(A,B),C)

  return OR(OR(OR(y1,y2),y3),y4)
```

Logical Expression for C-OUT:
= A B + A C-IN + B C-IN 

```
def COUT_FN(A,B,C):
  y1 = AND(A,B)
  y2 = AND(B,C)
  y3 = AND(A,C)

  return OR(OR(y1,y2),y3)
```
